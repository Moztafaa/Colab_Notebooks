```python
import requests
from bs4 import BeautifulSoup
import os
```

```python
# Replace with the archive.org URL of the Quran playlist
url = "https://archive.org/details/z240807_202408xxxxxx"
```

```python
# Folder where files will be saved
download_folder = '/content/drive/MyDrive/quran_downloads'
os.makedirs(download_folder, exist_ok=True)

# Start a session
session = requests.Session()
response = session.get(url)
```

```python
if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    # Find all download links
    links = soup.find_all('a', href=True)

    # Filter out only .mp3 or audio file links
    audio_links = [link['href'] for link in links if link['href'].endswith('.mp3')]

    for idx, audio_url in enumerate(audio_links):
        full_url = f"https://archive.org{audio_url}"
        file_name = os.path.join(download_folder, f"{idx + 1}.mp3")

        print(f"Downloading {file_name}...")

        # Download the file
        with session.get(full_url, stream=True) as file_response:
            with open(file_name, 'wb') as file:
                for chunk in file_response.iter_content(chunk_size=8192):
                    file.write(chunk)

        print(f"Downloaded {file_name}")

    print("All files downloaded successfully.")
else:
    print("Failed to retrieve page. Please check the URL.")
```
